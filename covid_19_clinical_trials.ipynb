## 4. data cleaning & Handle Missing Data

```python
print(df.isnull().sum())

# Drop columns with excessive missing data
df.drop(['Study Documents', 'Results First Posted'], axis=1, inplace=True)

# Fill missing categorical columns
df['Acronym'] = df['Acronym'].fillna('Missing Acronym')
df['Phases'] = df['Phases'].fillna('Missing Phases')
df['Interventions'] = df['Interventions'].fillna('Missing Interventions')
df['Locations'] = df['Locations'].fillna('Missing Location')
```

```python
# Fill numeric missing with median
df['Enrollment'] = df['Enrollment'].fillna(df['Enrollment'].median())

# Drop duplicates
df.drop_duplicates(inplace=True)
```

```python
# Feature: Extract Country from Locations
df['Country'] = df['Locations'].apply(lambda x: str(x).split(',')[-1] if pd.notnull(x) else 'Missing Country')

# Convert Start Date
df['Start Date'] = pd.to_datetime(df['Start Date'], errors='coerce')
```

## 5. EDA
### Univariate Analysis

```python
# Status distribution
df['Status'].value_counts().plot(kind='bar', title='Trial Status Distribution')
plt.show()
```

```python
# Phases distribution
df['Phases'].value_counts().plot(kind='bar', title='Trial Phases Distribution')
plt.show()
```

```python
# Top 10 countries
df['Country'].value_counts().head(10).plot(kind='bar', title='Top 10 Countries by Trial Count')
plt.show()
```

```python
# Age distribution
df['Age'].value_counts().plot(kind='bar', title='Age Group Distribution')
plt.show()
```

```python
# Gender distribution
df['Gender'].value_counts().plot(kind='bar', title='Gender Distribution')
plt.show()
```

## 6. Bivariate Analysis

```python
# Status vs. Phases:
pd.crosstab(df['Status'], df['Phases']).plot(kind='bar', stacked=True, title='Status vs. Phases')
plt.show()
```

```python
df['Start Date'].dt.to_period('M').value_counts().sort_index().plot()
plt.title('Trials Started Over Time')
plt.xlabel('Start Month')
plt.ylabel('Number of Trials')
plt.show()
```

```python
# Improved Start Month Analysis with Correct Visualization
import seaborn as sns
import matplotlib.pyplot as plt

# Remove rows where Start Date is missing
df_valid = df.dropna(subset=['Start Date'])

# Extract month numbers
start_month = df_valid['Start Date'].dt.month

# Count and plot
start_month_distribution = start_month.value_counts().sort_index()
print("Trials Started per Month:")
print(start_month_distribution)

# Bar plot using value_counts
plt.figure(figsize=(10,6))
start_month_distribution.plot(kind='bar')
plt.title('Distribution of Trials by Start Month')
plt.xlabel('Start Month')
plt.ylabel('Number of Trials')
plt.xticks(ticks=range(0,12), labels=['Jan','Feb','Mar','Apr','May','Jun','Jul','Aug','Sep','Oct','Nov','Dec'], rotation=45)
plt.show()
```

### more EDA

```python
sns.countplot(data=df, y='Status', hue='Gender')
plt.title('Trial Status by Gender')
plt.show()
```

```python
top_sponsors = df['Sponsor/Collaborators'].value_counts().head(20)
top_sponsors.plot(kind='barh', figsize=(10,8))
plt.title('Top 20 Sponsors/Collaborators')
plt.xlabel('Number of Trials')
plt.show()
```

```python
sns.histplot(df['Enrollment'], bins=50, kde=True)
plt.title('Enrollment Distribution')
plt.xscale('log')  # Many outliers, log scale works well
plt.show()
```

```python
df['Year'] = df['Start Date'].dt.year
df['Year'].value_counts().sort_index().plot(kind='bar')
plt.title('Number of Trials Started Each Year')
plt.xlabel('Year')
plt.ylabel('Number of Trials')
plt.show()
```

```python
top10 = df['Country'].value_counts().head(10).index

df_top10 = df[df['Country'].isin(top10)]

df_top10.groupby(['Year', 'Country']).size().unstack().plot(kind='line', marker='o', figsize=(14,7))
plt.title('Trials Started Over Years (Top 10 Countries)')
plt.xlabel('Year')
plt.ylabel('Number of Trials')
plt.legend(title='Country')
plt.show()
```

```python
# Example: Correlation heatmap for numeric features
import seaborn as sns
import matplotlib.pyplot as plt

# Calculate correlation matrix
corr = df.select_dtypes(include=['float64', 'int64']).corr()

# Plot heatmap
plt.figure(figsize=(10,8))
sns.heatmap(corr, annot=True, cmap='coolwarm', fmt='.2f', square=True)
plt.title('Correlation Heatmap')
plt.show()

# Heatmap of missing data
plt.figure(figsize=(12,6))
sns.heatmap(df.isnull(), cbar=False, yticklabels=False, cmap='viridis')
plt.title('Missing Data Heatmap')
plt.show()
```
